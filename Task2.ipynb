### Задача 1

Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. 
Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
p1 = 0.8
n1 = 100
k1 = 85
```

т.к. вероятность единичного события значительно больше 0.1, количество событий равно 100 и все события независимы друг от друга, то для определения вероятности используем формулу Бернулли

```python
# Функция для расчета числа сочетаний для n-случаев и k-исходов

from math import factorial as ftl

def combns(n,k):
    return int(ftl(n) / (ftl(k) * ftl(n-k)))
```

```python
# Расчет вероятности попадания в мишень 85 раз:

p_85 = combns(n1,k1) * p1**k1 * (1-p1)**(n1-k1)
print(str('Вероятность попадания в мишень 85 раз: %.4f' % p_85))
```

### Задача 3

Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

```python
p3 = 0.5 # вероятность выпадения орла за один бросок монеты
n3 = 144
k3 = 70
```

т.к. вероятность единичного события значительно больше 0.1, количество событий относительно невелико и все события независимы друг от друга, то для определения вероятности используем формулу Бернулли

```python
# Расчет вероятности выпадения орла 70 раз:

p_70 = combns(n3,k3) * p3**k3 * (1-p3)**(n3-k3)
print(str('Вероятность выпадения орла 70 раз: %.4f' % p_70))
```

### Задача 4

В первом ящике находится 10 мячей, из которых 7 - белые. 
Во втором ящике - 11 мячей, из которых 9 белых. 
Из каждого ящика вытаскивают случайным образом по два мяча. 
Какова вероятность того, что все мячи белые? 
Какова вероятность того, что ровно два мяча белые? 
Какова вероятность того, что хотя бы один мяч белый?

#### Условие 1: Все мячи белые

```python
# 1 ящик:
С2_7 = combns(7,2)   # число комбинаций взять 2 из 7 белых мяча
С2_10 = combns(10,2) # число комбинаций взять 2 любых мяча из ящика
p1b = С2_7 / С2_10   # вероятность взять 2 белых мяча из ящика

# 2 ящик:
С2_9 = combns(9,2)   # число комбинаций взять 2 из 9 белых мяча
С2_11 = combns(11,2) # число комбинаций взять 2 любых мяча из ящика
p2b = С2_9 / С2_11   # вероятность взять 2 белых мяча из ящика

print('Вероятность взять 2 белых шара из\n 1-го ящика: %.4f\n 2-го ящика: %.4f' % (p1b, p2b))

# Для определения общей вероятности частные вероятности перемножаются:
print('Вероятность взять по 2 белых шара из 2-х ящиков сразу: %.4f' % (p1b * p2b))
```

#### Условие 2: Два белых мяча 

Возможны комбинации:

   1. Событие А: Из 1-го ящика взяли 2 белых мяча, Событие В: из 2-го ящика взяли 2 мяча другого цвета;

   2. Событие С и D: Из обоих ящиков взяли по 1-му белому мячу и 1-му мячу другого цвета;

   3. Событие E: Из 1-го ящика взяли 2 мяча другого цвета, Событие F: из 2-го ящика взяли 2 белых мяча;

Общая вероятность события:
Р = p(A)p(B) + p(C)p(D) + p(E)p(F)

Число благоприятных исходов будет:

```python
# 1 комбинация:
m_A = combns(7,2)  # 2 мяча белого цвета из 7 мячей
m_B = combns(2,2)  # 2 мяча другого цвета из 11-9=2 мячей 

# 2 комбинация:
m_C = combns(7,1) * combns(3,1)  # 1 мяч белого цвета из 7 мячей и 1 мяч другого цвета из 10-7=3 мяча
m_D = combns(9,1) * combns(2,1)  # 1 мяч белого цвета из 9 мячей и 1 мяч другого цвета из 11-9=2 мяча

# 3 комбинация:
m_E = combns(3,2)  # 2 мяча другого цвета из 10-7=3 мячей
m_F = combns(9,2)  # 2 мяча белого цвета из 9 мячей

print(f'Число благоприятных исходов:\n m_A = {m_A}\n m_B = {m_B}\n m_C = {m_C}\n m_D = {m_D}\n m_E = {m_E}\n m_F = {m_F}')
```

Общая вероятность события:

```python
P=(m_A/С2_10)*(m_B/С2_11)+(m_C/С2_10)*(m_D/С2_11)+(m_E/С2_10)*(m_F/С2_11)
print('Вероятность взять 2 белых шара суммарно из 2-х ящиков: %.4f' % P)
```

#### Условие 3: Хотя бы один мяч белый

Наименьшее количество шагов расчета для этого условия будет, если рассмотреть обратную задачу: нам не нужен ни один белый мяч при взятии мячей из обоих ящиков. Соответственно мы рассматриваем полную группу событий, т.к. по прямому условию - один белый мяч должен быть взят, а по обратному условию - нет, т.е. сумма вероятностей по обоим условиям будет равна единице и тогда  P(A) = 1 - P(B).

Для обратного условия, когда не должно быть ни одного белого мяча:   

```python
# Число благоприятных событий при взятии мячей из обоих ящиков:
C3_2 = combns(3,2)  # 2 мяча другого цвета из 10-7=3 мячей
C2_2 = combns(2,2)  # 2 мяча другого цвета из 11-9=2 мячей
```

```python
P_B = C3_2 / С2_10 * C2_2 / С2_11
P_A = 1 - P_B

print('Вероятность когда нет ни одного белого мяча в событии: %.4f' % P_B)
print('Вероятность того, что хотя бы один белый мяч взят: %.4f' % P_A)
```
