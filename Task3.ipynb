### Задача 1

Даны значения зарплат из выборки выпускников:<br>
100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.<br> 
Посчитать без использования статистических методов наподобие std, var, mean:<br>
среднее арифметическое,<br>
среднее квадратичное отклонение,<br>
смещенную и несмещенную оценки дисперсий для данной выборки.

```python

zp = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

# Среднее арифметическое значение зарплаты:
mean = sum(zp) / len(zp)

# Среднее квадратичное отклонение:
qv = 0
for i in range(len(zp)):
    v = (zp[i]-mean)**2
    qv = qv + v
midqv = (qv / len(zp))**0.5

# Смещенная оценка дисперсии:
D = qv / len(zp)

# Несмещенная оценка дисперсии:
Dn = qv / (len(zp)-1)

print('Средняя зарплата, тыс.руб:\t%.2f' % mean)
print('Ср.квадратичное отклонение:\t%.2f' % midqv)
print('Смещенная оценка дисперсии:\t%.3f' % D)
print('Несмещенная оценка дисперсии:\t%.3f' % Dn)
```

т.к. вероятность единичного события значительно больше 0.1, количество событий равно 100 и все события независимы друг от друга, то для определения вероятности используем формулу Бернулли

### Задача 2

В первом ящике находится 8 мячей, из которых 5 - белые.<br>
Во втором ящике - 12 мячей, из которых 5 белых.<br>
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4.<br>
Какова вероятность того, что 3 мяча белые?

Возможны следующие комбинации мячей:<br>
 1. Событие А: из 1-го ящика взяли 2 белых мяча; <br>
    Событие В: из 2-го яшика взяли 1 белый мяч и 3 мяча другого цвета; <br><br>
 2. Событие С: из 1-го ящика взяли 1 мяч и 1 мяч другого цвета; <br>
    Событие D: из 2-го ящика взяли 2 белых мяча и 2 мяча другого цвета; <br><br>
 3. Событие E: из 1-го ящика взяли 2 мяча другого цвета; <br>
    Событие F: из 2-го ящика взяли 3 белых мяча и 1 мяч другого цвета. <br><br>
Общая вероятность события: Р = p(A)p(B) + p(C)p(D) + p(E)p(F)

```python
# Функция для расчета числа сочетаний для n-случаев и k-исходов

from math import factorial as ftl

def combns(n,k):
    return int(ftl(n) / (ftl(k) * ftl(n-k)))
```

```python
n1 = 8            # всего мячей в 1-м ящике
n2 = 12           # всего мячей во 2-м ящике 
nb = 5            # количество белых мячей по условию задачи в каждом ящике

s1 = [2, 1, 0]    # кол-во белых мячей взятых из 1-го ящика по событиям А, С, Е
s2 = [1, 2, 3]    # кол-во белых мячей взятых из 2-го ящика по событиям В, D, F
sr1 = s1[::-1]    # кол-во мячей другого цвета, взятых из 1-го ящика будет инверсно списку S1
sr2 = s2[::-1]    # кол-во мячей другого цвета, взятых из 2-го ящика будет инверсно списку S2

C2_8 = combns(8,2)    # число комбинаций взять любые два мяча из 1-й корзины
C4_12 = combns(12,4)  # число комбинаций взять любые четыре мяча из 2-й корзины
```

```python
# Рассчитаем вероятности по 3-м комбинациям случаев и определим общую вероятность события:
P = 0

for i in range(len(s1)):
    m1 = combns(nb, s1[i]) * combns((n1 - nb), sr1[i]) # число комбинаций взять мячи из 1-го ящика для i-го события 
    m2 = combns(nb, s2[i]) * combns((n2 - nb), sr2[i]) # число комбинаций взять мячи из 2-го ящика для i-го события
    P = P + m1 / C2_8 * m2 / C4_12                     # суммирование по событиям вероятностей взять белые мячи
    
print('Вероятность взять белые мячи по условию задачи:\t%.4f' % P)
```

### Задача 3

В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. <br>
Какова вероятность, что он учится: <br>
a). на факультете A <br>
б). на факультете B <br>
в). на факультете C?

```python
# Общее количество студентов университета примем за единицу, тогда вероятность, что студент с конкретного факультета:
p_B = [0.25, 0.25, 0.5]

# Вероятность сдачи сессии студентом конкретного факультета:
p_AB = [0.8, 0.7, 0.9]

f = ['A', 'B', 'C'] # список факультетов
```

```python
# Для расчета вероятности сдачи сессии применим формулу Байеса: 
#     P(B|A) = P(A|B)·P(B)/P(A),
# где P(A|B) - вероятность сдачи сессии студентом конкретного факультета; 
#     P(B) - вероятность, что студент учится на конкретном факультете; 
#     P(A) - общая вероятность сдачи сессии студентом любого факультета.

P_A = 0
for i in range(len(p_B)):
    P_A = P_A + p_B[i] * p_AB[i]  # Расчет общей вероятности P(A)

    
# Расчет вероятностей P(B|A) по условиям задачи:

print('Вероятность, что студент учится на факультете:')

for j in range(len(p_B)):
    P_BA = p_AB[j] * p_B[j] / P_A
    print(f' {f[j]}:\t{P_BA:.4}')
```

### Задача 4

Устройство состоит из трех деталей. <br>
Вероятность детали выйти из строя в первый месяц:<br>
&nbsp;&nbsp;первой&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;0.1<br>
&nbsp;&nbsp;второй&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;0.2<br>
&nbsp;&nbsp;третьей&nbsp;&nbsp;&nbsp;&nbsp;0.25<br>
Какова вероятность того, что в первый месяц выйдут из строя: <br>
&nbsp;&nbsp;а). все детали <br>
&nbsp;&nbsp;б). только две детали <br>
&nbsp;&nbsp;в). хотя бы одна деталь <br>
&nbsp;&nbsp;г). от одной до двух деталей?

```python
# Вероятность детали выйти из строя:
p1 = 0.1
p2 = 0.2
p3 = 0.25
```

###### а) Все выйдут из строя

```python
P1 = p1 * p2 * p3
print('Вероятность по условию a): %.4f' % P1)
```

###### б). Только две детали выйдут из строя

Общая вероятность сложится из трех вариантов развития событий:<br>
&nbsp;&nbsp;1.&nbsp;В&nbsp;В&nbsp;Н<br>
&nbsp;&nbsp;2.&nbsp;В&nbsp;Н&nbsp;В<br>
&nbsp;&nbsp;3.&nbsp;Н&nbsp;В&nbsp;В<br>
где "В" - выйдет из строя, "Н" - не выйдет из строя, т.е. вероятность будет обратной (1-р<sub>i</sub>).<br>
Поэтому:<br>  Р<sub>2</sub> = р<sub>1</sub>·р<sub>2</sub>·(1-р<sub>3</sub>) + р<sub>1</sub>·(1-р<sub>2</sub>)·р<sub>3</sub> + (1-р<sub>1</sub>)·р<sub>2</sub>·р<sub>3</sub>

```python
P2 = p1 * p2 * (1 - p3) + p1 * (1 - p2) * p3 + (1 - p1) * p2 * p3
print('Вероятность по условию б): %.4f' % P2)
```

###### в). Хотя бы одна деталь выйдет из строя

Проще рассмотреть обратную задачу, что ни одна деталь не вышла из строя, т.е. вероятность работоспособности будет 
р<sub>раб</sub> = (1-p<sub>1</sub>)(1-p<sub>2</sub>)(1-p<sub>3</sub>), соответственно хотя бы одна деталь выйдет из строя с обратной вероятностью 1-р<sub>раб</sub>.

```python
P3 = 1 - (1 - p1) * (1 - p2) * (1 - p3)
print('Вероятность по условию в): %.4f' % P3)
```

###### г). Выйдет из строя от одной до двух деталей

Вероятность выхода из строя хотя бы одной детали:<br>
Р = p<sub>1</sub>·(1-p<sub>2</sub>)(1-p<sub>3</sub>)+(1-p<sub>1</sub>)·p<sub>2</sub>·(1-p<sub>3</sub>)+(1-p<sub>1</sub>)(1-p<sub>2</sub>)·p<sub>3</sub><br>
Вероятность выхода из строя хотя бы двух деталей у нас посчитана в условии б, поэтому общая вероятность по условию:<br>
Р<sub>4</sub>=Р+Р<sub>2</sub>

```python
P = p1 * (1 - p2) * (1 - p3) + (1 - p1) * p2 * (1 -p3) + (1 - p1) * (1 - p2) * p3
P4 = P + P2
print('Вероятность по условию г): %.4f' % P4)
```
